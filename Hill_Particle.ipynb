###Discrete or particle based approach###

1. Transport of soil

```python
## x spatial grid for simulation and plotting
x_grid = np.linspace(-100,100,201) # unit meter grid

def getRandLocation(x_grid=x_grid):
    """returns a random index of the x_grid and return it"""
    return np.random.choice(len(x_grid))

def getRandDir():
    return np.random.choice(['L','R'])
```

###Probabilities to continue moving:###
   * 1 if drop
   * 7/8 if flat
   * 1/8 if up by one particle
   * 0 if up by two or more particles

```python
def moveDepositSed(elev_profile, psize, x_grid=x_grid):
    sedloc = sedinitloc = getRandLocation() ## initial location to keep track
    seddir = getRandDir() ## direction
    
    ## soil production may occur here, a check if bedrock if so, then call produce soil
    if elev_profile[sedloc] <= 0:
        return elev_profile
    
    moving = True         
    if seddir == 'R':
        while moving:
            ## check if on far right edge of profile, if so particle disappears to river
            if sedloc == len(x_grid)-1:  
                elev_profile[sedinitloc] -= psize 
                moving = False
                break
            ## check if particle is moving down a slope, if so move particle over by one
            if elev_profile[sedloc] - elev_profile[sedloc+1] > 0:
                sedloc += 1
                continue
            elif elev_profile[sedloc] - elev_profile[sedloc+1] <= -2*psize:
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break
            elif (elev_profile[sedloc] == elev_profile[sedloc+1]) and (np.random.rand() >= 7/8.):
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break  
            elif (elev_profile[sedloc] - elev_profile[sedloc+1] >= -psize) and (np.random.rand() >= 1/8.):
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break      
            else:
                sedloc += 1   
        return elev_profile
                
    elif seddir == 'L':
        while moving:            
            if sedloc == 0:
                elev_profile[sedinitloc] -= psize 
                moving = False
                break
            if elev_profile[sedloc] - elev_profile[sedloc-1] > 0:
                sedloc -= 1
                continue
            elif elev_profile[sedloc] - elev_profile[sedloc-1] <= -2*psize:
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                continue
            elif (elev_profile[sedloc] == elev_profile[sedloc-1]) and (np.random.rand() >= 7/8.):
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break   
            elif (elev_profile[sedloc] - elev_profile[sedloc-1] >= -psize) and (np.random.rand() >= 1/8.):
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break     
            else:
                sedloc -= 1   
        return elev_profile
```

```python
#######
## Parameters and boundary conditions
#######

## initial height from first uplift 
init_height = .15 
particle_size = .31 
uplift_height = .1 
## initial elevation profile
## boundary conditions are constant height of zero
elev = np.zeros_like(x_grid) + (init_height)
elev[0] = 0  
elev[len(x_grid)-1] = 0 

fig = plt.figure(figsize=(12,6))

for i in range(500000):
    elev = moveDepositSed(elev, particle_size)
    if i % 500 == 0:
        elev = elev + np.zeros_like(len(x_grid)) + (uplift_height)*np.random.rand()
        elev[0] = 0  
        elev[len(x_grid)-1] = 0 
    if i % 1000 == 0:
        #plt.plot(x_grid, elev, 'k', alpha = 0.7)
        plt.fill(x_grid, elev, 'green', alpha=0.5)
        plt.xlim(-100,100)
        
        display(plt.gcf())  
        clear_output()
```
