```python
######
## Particle based approach
######

## x spatial grid for simulation and plotting
x_grid = np.linspace(-100,100,201) # unit meter grid

## initial height from first uplift 
init_height = 10 

## initial elevation profile
elev = np.ones_like(x_grid) + np.random.rand(len(x_grid)) + (init_height -1)
## boundary conditions are constant height of zero
elev[0] = 0  
elev[len(x_grid)-1] = 0 

## plot initial uplift profile
plt.plot(x_grid, elev, alpha = 0.8)
plt.xlim(-110,110)
plt.show()
```

```python
def getRandLocation(x_grid=x_grid):
    """returns a random index of the x_grid and return it"""
    return np.random.choice(len(x_grid))

#### test if out of bounds
x = np.arange(10000)
y = np.array([np.random.rand() for i in x])

plt.scatter(x,y, s=.1)
print y.mean()
```

```python
def getRandDir():
    return np.random.choice(['L','R'])
```

```python
def moveDepositSed(elev_profile, psize, x_grid=x_grid):
    sedinitloc = getRandLocation()
    sedloc = sedinitloc
    seddir = getRandDir()
    
    moving = True    
    while moving:        
        ## to check if particle is on boundary, if so stop
        if (sedloc == len(x_grid)-1) or (sedloc == 0):
            moving = False
            break
                    
        elif seddir == 'R':
            while elev_profile[sedloc] - elev_profile[sedloc+1] > psize:
                sedloc += 1
                if sedloc == len(x_grid)-1:
                    elev_profile[sedinitloc] -= psize 
                    moving = False
                    break      
            elev_profile[sedloc] += psize
            elev_profile[sedinitloc] -= psize 
            moving = False         
                  
                
        elif seddir == 'L':
            while elev_profile[sedloc] - elev_profile[sedloc-1] > psize:
                sedloc -= 1
                if sedloc == 0:
                    elev_profile[sedinitloc] -= psize 
                    moving = False
                    break      
            elev_profile[sedloc] += psize
            elev_profile[sedinitloc] -= psize 
            moving = False  
                
    return elev_profile
```

```python
particle_size = 0.1
## initial elevation profile
elev = np.zeros_like(x_grid) + np.random.rand(len(x_grid)) + (init_height)
uplift_height = 0.1
## boundary conditions are constant height of zero
elev[0] = 0  
elev[len(x_grid)-1] = 0 


for i in range(500000):
    elev = moveDepositSed(elev, particle_size)
    if i % 1000 == 0:
        elev += np.random.rand(len(x_grid)) + (uplift_height)
        elev[0] = 0  
        elev[len(x_grid)-1] = 0 
    if i % 10000 == 0:
        plt.plot(x_grid, elev, alpha = 0.8)
        plt.xlim(-105,105)
```

Probabilities to continue moving:
    * 1 if drop
    * 7/8 if flat
    * 1/8 if up by one particle
    * 0 if up by two or more particles
