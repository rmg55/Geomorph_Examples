```python
######
## Particle based approach
######

## x spatial grid for simulation and plotting
x_grid = np.linspace(-50,50,101) # unit meter grid

## initial height from first uplift 
init_height = 10 

## initial elevation profile
elev = np.ones_like(x_grid) + np.random.rand(len(x_grid)) + (init_height -1)
## boundary conditions are constant height of zero
elev[0] = 0  
elev[len(x_grid)-1] = 0 

## plot initial uplift profile
plt.plot(x_grid, elev, alpha = 0.8)
plt.xlim(-110,110)
plt.show()
```

```python
def getRandLocation(x_grid=x_grid):
    """returns a random index of the x_grid and return it"""
    return np.random.choice(len(x_grid))

#### test if out of bounds
x = np.arange(10000)
y = np.array([np.random.rand() for i in x])
#plt.scatter(x,y, s=.1)
#print y.mean()
```

```python
def getRandDir():
    return np.random.choice(['L','R'])
```

Probabilities to continue moving:
    * 1 if drop
    * 7/8 if flat
    * 1/8 if up by one particle
    * 0 if up by two or more particles

```python
def moveDepositSed(elev_profile, psize, x_grid=x_grid):
    sedloc = sedinitloc = getRandLocation()
    #sedloc = sedinitloc
    seddir = getRandDir()
    
    if elev_profile[sedloc] <= 0:
        return elev_profile
    
    moving = True         
    if seddir == 'R':
        while moving:
            if sedloc == len(x_grid)-1:
                elev_profile[sedinitloc] -= psize 
                moving = False
                break
            if elev_profile[sedloc] - elev_profile[sedloc+1] > 0:
                sedloc += 1
                continue
            elif elev_profile[sedloc] - elev_profile[sedloc+1] <= -2*psize:
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break
            elif (elev_profile[sedloc] - elev_profile[sedloc+1] == -psize) and (np.random.rand() >= 1/8.):
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break      
            elif (elev_profile[sedloc] == elev_profile[sedloc+1]) and (np.random.rand() >= 7/8.):
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break  
            else:
                sedloc += 1
        return elev_profile
                
    elif seddir == 'L':
        while moving:            
            if sedloc == 0:
                elev_profile[sedinitloc] -= psize 
                moving = False
                break
            if elev_profile[sedloc] - elev_profile[sedloc-1] > 0:
                sedloc -= 1
                continue
            elif elev_profile[sedloc] - elev_profile[sedloc-1] <= -2*psize:
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                continue
            elif (elev_profile[sedloc] - elev_profile[sedloc-1] == -psize) and (np.random.rand() >= 1/8.):
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break     
            elif (elev_profile[sedloc] == elev_profile[sedloc-1]) and (np.random.rand() >= 7/8.):
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break   
            else:
                sedloc -= 1                         
        return elev_profile
```

```python
particle_size = 11 
## initial elevation profile
elev = np.zeros_like(x_grid) + (init_height)
uplift_height = 0.6 
## boundary conditions are constant height of zero
elev[0] = 0  
elev[len(x_grid)-1] = 0 


for i in range(500000):
    elev = moveDepositSed(elev, particle_size)
    if i % 10 == 0:
        elev = elev + np.zeros_like(len(x_grid)) + (uplift_height)
        elev[0] = 0  
        elev[len(x_grid)-1] = 0 
    if i % 10000 == 0:
        plt.scatter(x_grid, elev, s=0.5)
        plt.xlim(-50,50)
#plt.plot(x_grid, elev, alpha = 0.8)
```
