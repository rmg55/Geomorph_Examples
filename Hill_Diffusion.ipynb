##Continuous approach##

**1) Production of soil (from bedrock)**

Soil mass = regolith mass X production rate

Can use exponential decay with depth for soil production based on soil depth:

$$
\begin{align}
P &= P_0 e^{-kh} \\
\end{align}
$$

Where $P$ is the weathering rate, $P_0$ is the potential weathering rate, $k$ is an emprical constant, and $h$ is the soil thickness. 

Alternatively we can use the humped soil production function:

$$
\begin{align}
P &= P_0 \frac{h}{h_0} e^{-h/h_0} \\
\end{align}
$$

where $h_0$ is a constant.

**2) Transport of soil**

#### 1D Diffusion ####
$$\frac{\partial H}{\partial t} = -\frac{\partial Q}{\partial x} $$

and, 

$$Q = -k\frac{dH}{dx}, $$

when combined we get:

$$ \frac{\partial H}{\partial t} = k\frac{\partial^2H}{\partial x^2} $$

$Q=$ volumetric transport rate   
$H=$ height of soil 

#### Numerical solution using central differencing####

$$ \Delta H_x = k \frac{H_{x-1} -2H_{x} + H_{x+1} }{\Delta x^2} \Delta x $$

$$ H_{x,t} = H_{x,t-1} \Delta H_x $$

####Stability Criteria:####

$$ \Delta t < \frac{1}{2}\left(\frac{\Delta x^2}{k}\right) $$

```python
## x spatial grid for simulation and plotting
dx = 6 # delta x
x_grid = np.arange(-100,100,dx) 
## Initial conditions and parameters
R = np.zeros_like(x_grid) + 10.0
H = np.ones_like(x_grid) + R
R[0] = 0
R[len(R)-1] = 0
H[0] = 0
H[len(H) -1] = 0
k = .75
uplift_height = 0.01
tstart = 0
tend = 4000
numSteps = 4000
dt = (tend-tstart)/numSteps # delta t

## stability criteria
print 'dt =',dt ,'and must be less than =', 0.5*(dx**2/k)
```

```python
for i in np.arange(numSteps):  
    if i % 10 == 0:
        R = R + uplift_height
        H = H + uplift_height
        H = H - R
    #if i % 10 == 0:
        ns = 47 * np.exp(-0.02 * H) ## new soil from regolith
        R = R - ns  ## new bedrock 
        H = H + ns  ## new height of soil
    R[0] = 0
    R[len(R) -1] = 0
    H[0] = 0
    H[len(H)-1] = 0
    for j in np.arange(1,len(H)-1): 
        H[j] += k*dt*(H[j-1] - 2*H[j] + H[j+1])/(dx**2)  ## approximate second derivative
    if i % 20 == 0:
        plt.plot(x_grid,H)
        plt.xlim(-100,100)
        display(plt.gcf()) 
        clear_output()
plt.plot(x_grid,R)
```
