```python
## x spatial grid for simulation and plotting
x_grid = np.linspace(-100,100,201) # unit meter grid
```

```python
def getRandLocation(x_grid=x_grid):
    """returns a random index of the x_grid and return it"""
    return np.random.choice(len(x_grid))
```

```python
def getRandDir():
    return np.random.choice(['L','R'])
```

###Discrete or particle based approach###

1. Production of soil (from bedrock)
2. Transport of soil

###Soil production from bedrock weathering as exponential decay with soil depth###

$$
\begin{align}
P &= P_0 e^{-kh} \\
\end{align}
$$

Where $P$ is the weathering rate, $P_0$ is the potential weathering rate, $k$ is an emprical constant, and $h$ is the soil thickness. 

Alternatively we can use the humped soil production function, one example equation is:

$$
\begin{align}
P &= P_0 \frac{h}{h_0} e^{-h/h_0} \\
\end{align}
$$

where $h_0$ is a constant.


```python
h = np.arange(0,5,0.1)
P = 0.5* (h/0.7)*np.exp(-h/0.7)
plt.plot(h,P)
```

```python
def weather(bedrock_profile, elev_profile):
    soil_profile = elev_profile - bedrock_profile
    soil_profile[soil_profile <=0] = 0
    P = 0.5
    k = 0.5
    
    weathered = P * np.exp(-k * soil_profile)
    elev_profile += weathered
    bedrock_profile -= weathered
    #bedrock_profile[bedrock_profile <= 0] = 0
    
    return bedrock_profile, elev_profile
```

```python
def weatherHumped(bedrock_profile, elev_profile):
    soil_profile = elev_profile - bedrock_profile
    soil_profile[soil_profile <=0] = 0
    P = 0.5
    h = 0.7
    
    weathered = P * (soil_profile/h) * np.exp(-soil_profile/h)
    elev_profile += weathered
    bedrock_profile -= weathered
    #bedrock_profile[bedrock_profile <= 0] = 0
    
    return bedrock_profile, elev_profile
```

###Probabilities for particles to continue moving:###
   * 1 if drop
   * 7/8 if flat
   * 1/8 if up by one particle
   * 0 if up by two or more particles

```python
def moveDepositSed(elev_profile, bedrock_profile, psize, x_grid=x_grid):
    sedloc = sedinitloc = getRandLocation()
    seddir = getRandDir()
    
    ## a check if bedrock at surface if so, then no particle to move
    if elev_profile[sedloc] <= bedrock_profile[sedloc]:
        return elev_profile
    
    moving = True         
    if seddir == 'R':
        while moving:
            if sedloc == len(x_grid)-1:
                elev_profile[sedinitloc] -= psize 
                moving = False
                break
            if elev_profile[sedloc] - elev_profile[sedloc+1] > 0:
                sedloc += 1
                continue
            elif elev_profile[sedloc] - elev_profile[sedloc+1] <= -2*psize:
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break
            elif (elev_profile[sedloc] == elev_profile[sedloc+1]) and (np.random.rand() >= 7/8.):
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break  
            elif (elev_profile[sedloc] - elev_profile[sedloc+1] >= -psize) and (np.random.rand() >= 1/8.):
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break      
            else:
                sedloc += 1   
        return elev_profile
                
    elif seddir == 'L':
        while moving:            
            if sedloc == 0:
                elev_profile[sedinitloc] -= psize 
                moving = False
                break
            if elev_profile[sedloc] - elev_profile[sedloc-1] > 0:
                sedloc -= 1
                continue
            elif elev_profile[sedloc] - elev_profile[sedloc-1] <= -2*psize:
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                continue
            elif (elev_profile[sedloc] == elev_profile[sedloc-1]) and (np.random.rand() >= 7/8.):
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break   
            elif (elev_profile[sedloc] - elev_profile[sedloc-1] >= -psize) and (np.random.rand() >= 1/8.):
                elev_profile[sedloc] += psize
                elev_profile[sedinitloc] -= psize  
                moving = False
                break     
            else:
                sedloc -= 1 
        return elev_profile
```

```python
#######
## Parameters and boundary conditions
#######

## initial height for scarp
scarp_init = [10]*80 + [i+1 for i in np.arange(10,51,1)] + [51]*80 
particle_size = .31*2.75 
uplift_height = .06 
## initial elevation profile
## boundary conditions are constant height of zero
elev = np.zeros_like(x_grid) + (scarp_init)
bedrock = np.zeros_like(x_grid) + np.array(scarp_init)*0.95
elev[0] = 10  
elev[len(x_grid)-1] = 51 

fig = plt.figure(figsize=(12,8))

for i in range(100000):
    elev = moveDepositSed(elev, bedrock, particle_size)
    if i % 1000 == 0: #1500
        bedrock, elev = weatherHumped(bedrock, elev)
    if i % 1 == 0: #400
        #elev[0] = 10  
        elev[len(x_grid)-1] = 51 
    if i % 1000 == 0:
        #plt.plot(x_grid, elev, color='brown', alpha=.3)      
        plt.fill_between(x_grid, elev, bedrock, color='moccasin', alpha=.9)      
        plt.plot(x_grid, bedrock, 'darkslategray', alpha = 0.3)
        plt.fill_between(x_grid, bedrock,np.zeros_like(x_grid), color='darkslategray', alpha = 0.7)
        plt.xlim(-100,100)
        plt.ylim(0,np.max(elev)+10)
        display(plt.gcf()) 
        clear_output()
        
#plt.plot(x_grid, bedrock, 'darkslategray', alpha = 0.7, label='Bedrock')
#plt.legend()
```
